```python
# Dependencies
import numpy as np
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
```

```python
# Read in CSV.
df = pd.read_csv('Resources/cities.csv')
df.head()
```

```python
df.dtypes
```

```python
# Convert INT64 Date to Datetime
df['Date'] = pd.to_datetime(df['Date'], unit='s')
```

```python
df.head()
```

```python
# Convert to Date only.
df['Date'] = df['Date'].dt.date
```

```python
df = df.rename(columns={"Unnamed: 0": "City_ID", })
```

```python
df.head()
```

```python
# Output as HTML, drop index.
df.to_html('data_all.html', index=False)
```

```python
# Create scatter plot for Lat vs Max Temp.
plt.scatter(df['Lat'], df['Max Temp'], marker='o', s=20, c='b', edgecolors='green', alpha=.75)

# Format axis/labels.
plt.title('City Latitude vs. Max Temperature - 07/06/2019')
plt.ylabel('Max Temperature (F)')
plt.xlabel('Latitude')
plt.grid(True)
plt.tight_layout()

# Save the figure.
plt.savefig('Fig1.png')

plt.show()
```
